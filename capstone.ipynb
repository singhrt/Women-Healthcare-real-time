```python
##create dataframe of all na values
issDf=pd.DataFrame(whc.isnull().sum())
```

```python
###drop the complete columns where total na values are more than threshold...here more than 5000 out of 14644.
whc=whc.dropna(thresh=6000, axis=1)
whc.shape
```

```python
### here in train data..n and o values are numeric so mean method
#c values are character

for i in whc.columns.values:
    if i[0]=='n':
        whc[i]=whc[i].fillna(whc[i].mean())
    elif i[0]== ['c','o']:
        whc[i]=whc[i].fillna(whc[i].mode())
    else:
        pass
        
        
```
